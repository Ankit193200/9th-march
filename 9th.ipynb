# Q1: Probability Mass Function (PMF) and Probability Density Function (PDF):

Probability Mass Function (PMF):
The PMF is a function that gives the probability that a discrete random variable is exactly equal to some value. For example, let's consider the roll of a fair six-sided die. The PMF for this scenario would be a function that assigns probabilities to each possible outcome (1, 2, 3, 4, 5, 6).

For a fair die, the PMF would be P(X = x) = 1/6 for each x in {1, 2, 3, 4, 5, 6}.

Probability Density Function (PDF):
The PDF is used for continuous random variables. It represents the likelihood of a variable falling within a particular range. Unlike the PMF, which deals with discrete outcomes, the PDF deals with continuous outcomes.

An example could be the height of people in a population. The PDF would provide the probability that a person's height falls within a specific range, say between 160 cm and 170 cm.

# Q2: Cumulative Density Function (CDF):

The Cumulative Distribution Function (CDF) is the probability that a random variable takes a value less than or equal to a given point. It is the cumulative sum (or integral, in the case of continuous distributions) of the probability density function.

Example: For a standard normal distribution, the CDF at a value 'x' gives the probability that a randomly selected observation from that distribution is less than or equal to 'x'.

CDF is useful as it provides a way to determine the probability of a random variable falling within a certain range.

# Q3: Normal Distribution:

Normal distribution might be used as a model in situations where the data tends to cluster around a central value with no bias to the left or right. Examples include height or weight distributions in a population, IQ scores, errors in measurements, etc.

The parameters of the normal distribution are mean (μ) and standard deviation (σ). The mean determines the center of the distribution, and the standard deviation determines the spread. The larger the standard deviation, the more spread out the distribution.

# Q4: Importance of Normal Distribution:

Normal distribution is important because of the Central Limit Theorem, which states that the distribution of the sum (or average) of a large number of independent, identically distributed random variables approaches a normal distribution, regardless of the original distribution. This makes it a key concept in statistical inference.

Real-life examples include IQ scores, blood pressure measurements, heights of a population, errors in measurements, etc.

# Q5: Bernoulli Distribution:

The Bernoulli distribution is a discrete probability distribution for a random variable that can take on one of two possible outcomes, usually labeled as 0 and 1, with probabilities of success (1) and failure (0).

Example: Tossing a coin where 1 represents "heads" and 0 represents "tails".

Difference: Bernoulli Distribution deals with a single trial, whereas the Binomial Distribution deals with multiple independent, identical Bernoulli trials.

# Q7: Uniform Distribution:

Uniform Distribution is a probability distribution where every possible value of the random variable has an equal likelihood of occurring.

Example: Rolling a fair six-sided die. Each number (1, 2, 3, 4, 5, 6) has an equal probability of 1/6.

# Q8: Z Score:

The Z score (or standard score) measures how many standard deviations a data point is from the mean. It's calculated as 
Z= (X−μ)/σ


Importance: Z scores help in standardizing values, making it easier to compare different datasets. They are crucial in statistical analysis, especially in normal distribution analysis and hypothesis testing.

# Q9: Central Limit Theorem (CLT):

The Central Limit Theorem states that the distribution of the sum (or average) of a large number of independent, identically distributed random variables approaches a normal distribution, regardless of the original distribution. It is significant because it allows for the use of normal distribution-based statistical methods even when the underlying data may not be normally distributed.

# Q10: Assumptions of Central Limit Theorem:

Independence: The observations must be independent.
Sample Size: The sample size should be sufficiently large. There isn't a strict cutoff, but a common rule of thumb is a sample size greater than 30.
Identical Distribution: The random variables should be identically distributed, meaning they have the same mean and standard deviation.
